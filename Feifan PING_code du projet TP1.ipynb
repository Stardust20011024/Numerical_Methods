## Approximation de l'équation de transport à vitesse constante

Soit un temps final $T > 0$, nous cherchons une approximation de, $\overline{u}: \mathbb{R} \times (0,T) \rightarrow \mathbb{R}$, solution du problème suivants,  

$$
\left\{
\begin{array}{l}
\partial_t \overline{u}(x,t) + a \partial_x \overline{u}(x,t) = 0, \ \ \ \ \ \ \ \    
\forall (x,t) \in \mathbb{R} \times (0,T),\\
\overline{u}(x,0) = u_{ini}(x), \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \      
\forall x \in \mathbb{R},
\end{array}
\right.
$$

On rappelle que la solution de (P) est donnée par $\overline{u}(x,t)=u_{ini}(x-at), \forall (x,t) \in \mathbb{R} \times (0,T)$. Afin de travailler sur une domaine d'espace de dimension finie, on prendra une condition initiale $u_{ini}$ 1-périodique  

-------------------------------------------------------------------------------------------------------

**La démonstration de** $\overline{u}(x,t) = u_{ini} (x-at)$:

Puisque le problème donne que:  

$$
\left\{
\begin{array}{l}
\partial_t \overline{u}(x,t) + a \partial_x \overline{u}(x,t) = 0,    
\forall (x,t) \in \mathbb{R} \times (0,T),\\
\overline{u}(x,0) = u_{ini}(x),         
\forall x \in \mathbb{R},
\end{array}
\right.
$$

Nous considérons x comme une fonction de t, donc on peut obtenir que:  

$$
\left\{
\begin{array}{l}
x'(t) = a\\
x(0) = x_0
\end{array}
\right.
$$

En résolvant le système d'équations ci-dessus, on a la relation entre x and t: $x(t) = at + x_0$  
C'est-à-dire $x_0 = x - at$  
Donc pour tout $x_0 \in \mathbb{R}$, $\overline{u}(x(t),t) = u_{ini}(x_0) = u_{ini}(x - at)$, $\forall (x,t) \in \mathbb{R} \times (0,T)$

### Exercice 1 
Montrer que, puisque $u_{ini}$ est 1-périodique, alors $\overline{u}$ est également 1-périodique en espace, i.e. $\overline{u}(x+1,t)=\overline{u}(x,t),\  \forall (x,t) \in \mathbb{R} \times (0,T)$.

---------------------------------------------------------------------

**Démonstration:**  
$u_{ini}$ est 1-périodique implique que  
$$\forall x \in \mathbb{R}, u_{ini}(x+1) = u_{ini}(x)$$  
Puisque $\overline{u}(x,t) = u_{ini}(x-at)$  
Donc on a que  
$$\overline{u}(x+1,t) = u_{ini}((x+1)-at) = u_{ini}((x-at)+1) = u_{ini}(x-at) = \overline{u}(x,t)$$  
Donc $\overline{u}$ est également 1-périodique en espace

On définit une solution numérique discrète de l'équation comme une suite de valeurs $u_j^n$ à double indice approchant la solution exacte  $\overline{u}(x,t)$ au sens où $u_j^n \approx \overline{u} (x_j,t_n)$ avec $x_j = j \Delta x$ et $t_n = n \Delta t$, pour $j \in \mathbb{Z}$ et $n \in \mathbb{N}$. On nomme $\Delta x$ le pas d'espace et $\Delta t$ le pas de temps. On pose $\lambda = \frac{\Delta t}{\Delta x}$.  

Commençons par quelques exemples de schémas explicites pour l'équation de transport.

---------------------------------------------------------------------

#### Schéma centré.
Ce schéma est obtenu en utilisant une approximation centrée de la dérivée spatiale.

$$
u_j^{n+1} = u_j^n - \frac{\lambda a}{2} (u_{j+1}^n - u_{j-1}^n).
$$

---------------------------------------------------------------------

**Démonstration de la forme duschéma centré:** 

On va utiliser la formule de Taylor:  
Sur le temps la formule de Taylor d'ordre 1 sur le point $(x_j,t_{n+1})$ donne que:

$$
\overline{u}(x_j,t_{n+1}) = \overline{u}(x_j,t_n) + \partial_t \overline{u}(x_j,t_n)\Delta t\ + O(\Delta t)
$$

Donc on peut approcher $\partial_t \overline{u}$ par:  

$$
\partial_t \overline{u}(x_j,t_n) \approx \frac{\overline{u}(x_j,t_n + \Delta t) - \overline{u}(x_j,t_n)}{\Delta t}
$$

On dit qu'on applique ***le forward method*** sur le temps, qui est d'ordre 1. 

Sur l'espace la formule de Taylor d'ordre 2 sur le point $(x_{j+1},t_n)$ donne que:

$$
\overline{u}(x_{j+1},t_n) = \overline{u}(x_j,t_n) + \Delta t \partial_x \overline{u}(x_j,t_n) + \frac{\Delta x^2}{2} \partial_{xx}^{2} \overline{u}(x_j,t_n) + O(\Delta x^2)\\
\overline{u}(x_{j-1},t_n) = \overline{u}(x_j,t_n) - \Delta t \partial_x \overline{u}(x_j,t_n) + \frac{\Delta x^2}{2} \partial_{xx}^{2} \overline{u}(x_j,t_n) + O(\Delta x^2)
$$

On ajoute les deux équations et après on va obtenir que:

$$
\overline{u}(x_{j+1},t_n) - \overline{u}(x_{j-1},t_n) = 2 \Delta x \partial_x \overline{u}(x_j,t_n) + O(\Delta x^2)
$$

Donc on peut approcher $\partial_x \overline{u}(x_j,t_n)$ par:

$$
\partial_x \overline{u}(x_j,t_n) \approx \frac{\overline{u}(x_{j+1},t_n) - \overline{u}(x_{j-1},t_n)}{2\Delta x}
$$

On dit qu'on applique ***le center method*** sur l'espace, qui est d'ordre 2.  

On remplace $\partial_t \overline{u}(x_j,t_n)$ et $\partial_x \overline{u}(x_j,t_n)$ par les rèrultats obtenues avant, donc on a que:

$$
\frac{1}{\Delta t}(u_j^{n+1} - u_j^n) + \frac{a}{2\Delta x}(u_{j+1}^n - u_{j-1}^n) = 0
$$

Nous va mettre la partie de temps de (n+1) à gauche et la partie de temps de n à droit, après on obtient finalement la formule de shéma centré:

$$
u_j^{n+1} = u_j^n - \frac{a\lambda}{2} (u_{j+1}^n - u_{j-1}^n)  \ \ \ \ \ \ \ \ \ avec \lambda = \frac{\Delta t}{\Delta x}
$$

---------------------------------------------------------------------

#### Schéma décentré à gauche.
Il est obtenu en utilisant une approximation décentrée(à gauche) de la dérivée spatiale.

$$
u_j^{n+1} = u_j^n - \lambda a (u_j^n - u_{j-1}^n).
$$

---------------------------------------------------------------------

**Démonstration de la forme du schéma décentré à gauche:**

On applique le ***forward method*** sur le temps, donc on a que:  

$$
\partial_t u_j^n = \frac{u_j^{n+1} - u_j^n}{\Delta t}
$$

Sur l'espace, on utilise la formule de Taylor d'ordre 1 sur le point $(x_{j-1},t_n)$:  

$$
\overline{u}(x_{j-1},t_n) = \overline{u}(x_j,t_n) - \Delta x \partial_x \overline{u}(x_j,t_n) + O(\Delta x)
$$

Donc on peut approcher $\partial_x \overline{u}$ par le ***backward method***:  

$$
\partial_x u_j^n = \frac{u_j^n - u_{j-1}^n}{\Delta x}
$$

On remplace $\partial_t \overline{u}(x_j,t_n)$ et $\partial_x \overline{u}(x_j,t_n)$ par les rèrultats obtenues avant, donc on a que:

$$
\frac{1}{\Delta t}(u_j^{n+1} - u_j^n) + \frac{a}{\Delta x}(u_j^n - u_{j-1}^n) = 0
$$

Nous voulons mettre la partie de temps de (n+1) à gauche et la partie de temps de n à droit, après on obtient finalement la formule de shéma décentré à gauche:

$$
u_j^{n+1} = u_j^n - \lambda a(u_j^n - u_{j-1}^n) \ \ \ \ \ \ \ \ avec \  \lambda = \frac{\Delta t}{\Delta x}
$$

---------------------------------------------------------------------

#### Schéma décentré à droite.
Il est obtenu en utilisant une approximation décentrée(à droite) de la dérivée spatiale.

$$
u_j^{n+1} = u_j^n - \lambda a (u_{j+1}^n - u_j^n).
$$

---------------------------------------------------------------------

**Démonstration de la forme du schéma décentré à droite:**

On applique le ***forward method*** sur le temps, donc on a que:  

$$
\partial_t u_j^n = \frac{u_j^{n+1} - u_j^n}{\Delta t}
$$

Sur l'espace, on utilise la formule de Taylor d'ordre 1 sur le point $(x_{j+1},t_n)$:  

$$
\overline{u}(x_{j+1},t_n) = \overline{u}(x_j,t_n) + \Delta x \partial_x \overline{u}(x_j,t_n) + O(\Delta x)
$$

Donc on peut approcher $\partial_x \overline{u}$ par le ***forward method***:  

$$
\partial_x u_j^n = \frac{u_{j+1}^n - u_j^n}{\Delta x}
$$

On remplace $\partial_t \overline{u}(x_j,t_n)$ et $\partial_x \overline{u}(x_j,t_n)$ par les rèrultats obtenues avant, donc on a que:

$$
\frac{1}{\Delta t}(u_j^{n+1} - u_j^n) + \frac{a}{\Delta x}(u_{j+1}^n - u_j^n) = 0
$$

Nous va mettre la partie de temps de (n+1) à gauche et la partie de temps de n à droit, après on obtient finalement la formule de shéma décentré à gauche:

$$
u_j^{n+1} = u_j^n - \lambda a(u_{j+1}^n - u_j^n) \ \ \ \ \ \ \ \ avec \  \lambda = \frac{\Delta t}{\Delta x}
$$

---------------------------------------------------------------------

#### Schéma de Lax-Friedrichs.
On reconnaît une modification du schéma centré dans lequel $u_j^n$ est remplacé par la moyenne de $u_{j-1}^n$ et $u_{j+1}^n$.

$$
u_j^{n+1} = \frac{1}{2} (u_{j+1}^n + u_{j-1}^n) - \frac{\lambda a}{2}(u_{j+1}^n - u_{j-1}^n).
$$

---------------------------------------------------------------------

**Démonstration de la forme du schéma de Lax-Friedrichs:**

Par schéma centré, on a que:  

$$
u_j^{n+1} = u_j^n - \frac{\lambda a}{2} (u_{j+1}^n - u_{j-1}^n)  \ \ \ \ \ \ \ \ \ avec \lambda = \frac{\Delta t}{\Delta x}
$$

on va remplacer $u_j^n$ par la moyenne de $u_{j+1}^n$ et $u_{j-1}^n$, c'est-à-dire $u_j^n = \frac{1}{2}(u_{j+1}^n + u_{j-1}^n)$, donc on peut obtenir le schéma de Lax-Friedrichs:

$$
u_j^{n+1} = \frac{1}{2}(u_{j+1}^n + u_{j-1}^n) - \frac{\lambda a}{2}(u_{j+1}^n - u_{j-1}^n)
$$

---------------------------------------------------------------------

#### Schéma de Lax-Wendroff.
On reconnaît une modification du schéma centré dans lequel on a rajouté un terme étrange qui contient une discrétisation de la dérivée seconde en espace alors qu'il n'y a pas de dérivée seconde dans l'EDP considérée.

$$
u_j^{n+1} = u_j^n - \frac{\lambda a}{2}(u_{j+1}^n - u_{j-1}^n) + \frac{\lambda^2 a^2}{2}(u_{j+1}^n - 2u_j^n + u_{j-1}^n).
$$

---------------------------------------------------------------------

**Démonstration de la forme du schéma de Lax-Wendroff:**

Par la formule de Taylor d'ordre 2 sur le point $(x_j,t_{n+1})$, on a que:

$$
u_j^{n+1} = u_j^n + \Delta t \partial_t u_j^n + \frac{\Delta t^2}{2} \partial_{tt}^2 u_j^n
$$

Par le problème de cette exercice, on a que:

$$
\partial_t \overline{u}(x,t) + a\partial_x \overline{u}(x,t) = 0
$$

Donc on peut obtenir que:

$$
\partial_t u_j^n = - a\partial_x u_j^n \\
\partial_{tt}^2 u_j^n = \partial_t(-a\partial_x u_j^n) = -a\partial_x(\partial_t u_j^n) = a^2 \partial_{xx}^2 u_j^n
$$

Par formule de Taylor d'ordre 2 sur l'espace:

$$
u_j^{n+1} = u_j^n + \Delta x \partial_x u_j^n + \frac{\Delta x^2}{2} \partial_{xx}^2u_j^n + O(\Delta x^2)\\
u_j^{n-1} = u_j^n - \Delta x \partial_x u_j^n + \frac{\Delta x^2}{2} \partial_{xx}^2u_j^n + O(\Delta x^2)
$$

Donc on peut obtenir le ***center method d'ordre 2*** que:

$$
\partial_{xx}^2 u_j^n = \frac{1}{\Delta x^2}(u_{j+1}^n - 2u_j^n + u_{j-1}^n)\\
\partial_{tt}^2 u_j^n = a^2 \partial_{xx}^2 u_j^n = \frac{a^2}{\Delta x^2}(u_{j+1}^n - 2u_j^n + u_{j-1}^n)
$$

On utilise le ***center method*** pour approcher $\partial_x u_j^n$:

$$
\partial_x u_j^n = \frac{1}{2\Delta x}(u_{j+1}^n - u_{j-1}^n)
$$

Donc on remplace $\partial_t u_j^n$ et $\partial_{tt}^2 u_j^n$ par les résultats obtenus avant dans la formule de Taylor d'ordre 2 sur le point $(x_j,t_{n+1})$, c'est justement le schéma de Lax-Wendroff:

$$
u_j^{n+1} = u_j^n - \frac{\lambda a}{2}(u_{j+1}^n - u_{j-1}^n) + \frac{\lambda^2 a^2}{2}(u_{j+1}^n - 2u_j^n + u_{j-1}^n) \ \ \ \ \ \ \ \ avec \lambda = \frac{\Delta t}{\Delta x}
$$


---------------------------------------------------------------------

On peut les écrire sous la forme

$$
u_j^{n+1} = H(u_{j-L}^n,....,u_{j+L}^n) = \sum_{l=-L}^{+L}c_l u_{j+l}^n  
$$

où on a introduit une fonction $H$ appelée "solution discrète". Tous les schémas présentés sont des schémas à 3 points, c'est-à-dire qu'ils n'utilisent qu'au plus, les valeurs aux trois points $x_{j-1},\ x_j$ et $x_{j+1}$ pour avancer en temps (sous la forme, $L = 1$).

On va écrire la procédure d'itération sous la forme de matrice:  

$$
U_h^{n+1} = Q U_h^n
$$

Où on peut préciser la matrice $Q \in M_J(\mathbb{R})$ est de la forme:

$$
Q=
\begin{pmatrix}
c_0 & c_1 & 0 & 0 & ... & 0 & 0 & 0 & c_{-1}\\
c_{-1} & c_0 & c_1 & 0 & ... & 0 & 0 & 0 & 0\\
0 & c_{-1} & c_0 & c_1 & ... & 0 & 0 & 0 & 0\\
... & ... & ... & ... & ... & ... & ... & ... & ...\\
0 & 0 & 0 & 0 & ... & c_{-1} & c_0 & c_1 & 0\\
0 & 0 & 0 & 0 & ... & 0 & c_{-1} & c_0 & c_1\\
c_1 & 0 & 0 & 0 & ... & 0 & 0 & c_{-1} & c_0\\
\end{pmatrix}
_{J \times J}
$$

Pour obtenir la solution numérique quand $t=t_M$, il faut faire l'itération M fois à partir de la condition initiale.

```python
def schema(h_x, h_t, X, M, c_n1, c_0, c_p1):    #### n:négative ; p:positive
    
    t_M = M * h_t                               #### le temps final de simulation
    
    
    
    ### Tout d'abord on va calculer la solution exacte quand le temps t égale t_M 
    
    U_exact_M = []                              #### liste pour le stockage des solutions exactes quand t = t_M
    for x_j in X: 
        u_exact_M_j = u_ini(x_j - a*t_M)        #### le résultat de la première démonstration
        U_exact_M.append(u_exact_M_j)
    
    
    
    ### Ensuite on va calculer la solution numérique quand le temps t égale t_M
    
    U_num_0 = []                                #### liste pour le stockage des solutions exactes quand t = t_0
    for x_j in X:
        u_0_j = u_ini(x_j)
        U_num_0.append(u_0_j)
    
    mainline = np.ones(J)
    upline = np.ones(J-1)
    downline = np.ones(J-1)
    
    Q_main = np.diag(mainline)*c_0
    Q_up = np.diag(upline*c_p1, k=1)
    Q_down = np.diag(downline*c_n1, k=-1)
    Q = Q_main + Q_up + Q_down
    Q[0, -1] = c_n1
    Q[-1, 0] = c_p1
    
    U_numerique_M = U_num_0                     #### liste pour le stockage des solutions numériques quand t = t_M
    index = 1
    while index <= M:                           #### pour faire l'itération M fois
        U_numerique_M = np.dot(Q, U_numerique_M)
        index += 1
        
    return U_exact_M, U_numerique_M, Q
    
```

## Analyse des schémas
L'analyse numérique des schémas a pour objectif d'étudier leur convergence, dans une norme appropriée, et lorsqu'ils convergent, d'estimer l'erreur, calculée(souvent) dans la même norme, entre la solution exacte et la solution approchée. Comme nous l'avons déja vu pour l'équation de la chaleur, deux propriétés importantes entrent en jeu: la stabilité et la consistance, dont on déduit la convergence.

**Quelques notes**

On a déja vu toutes les cinq schémas pour approcher la solution analytique dans partie 1.  
Dans partie 2, notre objet est justifier si chaque schéma est un bon schéma pour l'approximation.  
C'est-à-dire on va étudier la propriété de convergence, donc la stabilité et la consistance.

---------------------------------------------------------------------

**Définition 2.1**
*On appelle erreur de consistance du schéma(1) au point $x_j$ et à l'instant $t_n$, le réel*

$$
\mathcal{k}_j^n = \frac{1}{\Delta t} (u(x_j,t_{n+1}) - H(u(x_{j-L},t_n),...,u(x_{j+L},t_n))).
$$

*L'erreur de consistance du schéma à l'instant $t_n$ est le vecteur $\mathcal{K}_h^n$ dont les composantes sont les $\mathcal{k}_j^n$, et qui vérifie donc*

$$
\overline{U}_h^{n+1} = Q \overline{U}_h^n + \Delta t \mathcal{K}_h^n
$$

---------------------------------------------------------------------

**Définition 2.2** On dit qu'un schéma est stable pour une norme vectorielle $||\ ||$ si et seulement si il existe une constante $C_0$ (qui peut dépendre de $T$), telle que

$$
\underset{n\Delta t \leq T}{sup} ||Q^n|| \leq C_0
$$

### Exercice 4
Nous allons étudier la stabilité du schéma. Calculer numériquement la norme matricielle pour chaque schéma. **Dans ce cas, on prend** $T=0.75,\ a=1,\ J=20,\ \lambda = 0.8$. Qu'observez-vous? Ensuite, trouver la condition écrivant à la main pour chaque schéma(si possible) afin de déterminer sa stabilité.

*Tout d'abord, on va calculer la norme matricielle pour chaque schéma.*

```python
#### définir une fonction pour calculer la norme matricielle d'ordre 2 ou infini de Ah pour chaque schéma avant

def norm_schema(type_schema, type_norm, J, Lambda): 
    
    c_n1, c_0, c_p1 = schema_type(type_schema)
    h_x, h_t, X, M = espaceDiscretise(J, Lambda)
    Q = schema(h_x, h_t, X, M, c_n1, c_0, c_p1)[2]
    norm_Q_m = []
    index = 1
    
    
    if type_norm == '2':
        norm_Q = np.linalg.norm(Q, ord=2)
        while index <= M:
            norm = np.linalg.norm(np.matrix(Q)**index, ord=2)
            norm_Q_m.append(norm)
            index += 1
        norm_Q_M = np.max(norm_Q_m)
    elif type_norm == 'infini':
        norm_Q = np.linalg.norm(Q, ord=np.inf)
        while index <= M:
            norm = np.linalg.norm(np.matrix(Q)**index, ord=np.inf)
            norm_Q_m.append(norm)
            index += 1
        norm_Q_M = np.max(norm_Q_m)
    else:
        return print("Ca fonction ne peut pas calculer votre demande")
    
    return norm_Q, norm_Q_M
    
```

***Quelques notes sur la condition de stabilité pour chaque schéma:***
* Il y a deux schémas non stables pour les deux normes matricielles sous les conditions données, la norme matricielle $||Q^n||$ n'est pas bornée pour $|| \ \ ||_2$ et $||\ \  ||_{\infty}$:  
  Pour schéma centré, les coefficients ne sont pas tous positives. Cela vérifie le résultalt du cours que "schéma centré s'avère numériquement instable" n'importe quelle condition.    
  Pour schéma décentré à gauche, puisque ici $a=1>0$ qui est contre la condition stable "$a<0,\ -a\lambda \leq 1$" du cours.  
* Il y a deux schémas stables pour les deux normes matricielles sous les conditions données(schéma décentré à gauche et schéma Lax-Friedrichs), la norme matricielle est à valeur vers 1.   
  Pour schéma décentré à gauche, $c_{-1}=a\lambda=0.8\geq 0;\ c_0=1-a\lambda=0.2\geq 0;\ c_1 = 0$  
  Pour schéma Lax-Friedrichs, $c_{-1}=\frac{1}{2}+\frac{a\lambda}{2}=0.9\geq 0;\ c_0=0;\ c_1 = \frac{1}{2}-\frac{a\lambda}{2}=0.1\geq 0$  
   Du cours, ils sont vérifié la condition CFL:$|a|\lambda \leq 1$ donc stables.  
* Il y a un schéma stable par rapport à la norme matricielle de la norme 2 mais pas stable par rapport à la norme $\infty$(schéma Lax-Wendroff). Donc quand on choisit ce schéma pour calculer, on doit bien choisir la norme pour vérifier la stabilité.

### Exercice 5
Nous devons étudier la convergence pour chaque schéma. On calcul la vitesse à laquelle l'erreur décroît lorsqu'on diminue $\Delta x$(en aumentant $J$, le nombre des points). On dit que la méthode est d'ordre $p\in \mathbb{R}^{+}$ s'il existe une constante $C>0$, indépendante de $\Delta x$, telle que $||\overline{U}_{\Delta x}^n - U_{\Delta x}^n|| = \mathcal{O}((\Delta x)^p).$ On utilisera les normes $L^q$ discrète. Pour différents valeurs de $\Delta x$, calculer les erreurs $||\overline{U}_{\Delta x}^n - U_{\Delta x}^n||_{2,\Delta}$ et $||\overline{U}_{\Delta x}^n - U_{\Delta x}^n||_{\infty,\Delta}$. \epsilon_{\Delta t,\Delta x}^{(\infty)} = \underset{n \geq 0}{max}(||\overline{U}_{\Delta x}^n - U_{\Delta x}^n||_{\infty ,\Delta})\\

où $U_{\Delta x}^n = (u_j^n)_j$ et $\overline{U}_{\Delta x}^n = (\overline{u}(x_j, t_n))_j$. **Dans ce cas, on prend** $T=0.75,\ a=1,$$J=[25,50,100,200],\ \lambda = 0.8$. Qu'en observez-vous?

**Quelques notes**  

Dans cette exercice, on s'intéresse à la relation entre les erreurs et les valeurs de $\Delta x$ qu'on prend.  
Donc tout d'abord, il faut définir une fonction pour calculer les normes $L^2$ et $L^{\infty}$ d'erreur.  
En suite, il faut prendre les différents valeurs de $\Delta x$ et calculer les normes d'erreur.  
Pour montrer plus clairement, on prends la fonction log dans l'image pour montrer la relation entre eux.

```python
def schema_erreur(h_x, h_t, X, M, c_n1, c_0, c_p1, J):    #### n:négative ; p:positive
    
    t_M = M * h_t                                         #### le temps final de simulation
    
    
    
    ### Tout d'abord on va calculer la solution exacte quand le temps t égale t_M 
    
    U_exact_M = []                                        #### liste pour le stockage des solutions exactes quand t = t_M
    for x_j in X: 
        u_exact_M_j = u_ini(x_j - a*t_M)                  #### le résultat de la première démonstration
        U_exact_M.append(u_exact_M_j)
    
    ### Ensuite on va calculer la solution numérique quand le temps t égale t_M
    
    U_num_0 = []                                          #### liste pour le stockage des solutions exactes quand t = t_0
    for x_j in X:
        u_0_j = u_ini(x_j)
        U_num_0.append(u_0_j)
    
    mainline = np.ones(J)
    upline = np.ones(J-1)
    downline = np.ones(J-1)
    
    Q_main = np.diag(mainline)*c_0
    Q_up = np.diag(upline*c_p1, k=1)
    Q_down = np.diag(downline*c_n1, k=-1)
    Q = Q_main + Q_up + Q_down
    Q[0, -1] = c_n1
    Q[-1, 0] = c_p1
    
    U_numerique_M = U_num_0                               #### liste pour le stockage des solutions numériques quand t = t_M
    liste_erreur_inf = []                                 #### liste pour le stockage de norm inf d'erreur entre solutions numériques et solutions exactes pour chaque fois d'itération
    liste_erreur_2 = []                                   #### liste pour le stockage de norm 2 d'erreur entre solutions numériques et solutions exactes pour chaque fois d'itération
    
    
    index = 1
    while index <= M:                                     #### pour faire l'itération M fois
        t_m = index*h_t
        U_numerique_M = np.dot(Q, U_numerique_M)
        U_exact_M = u_ini(X - a*t_m)
        erreur_inf_m = np.linalg.norm(U_numerique_M - U_exact_M, np.inf)
        erreur_2_m = np.linalg.norm(U_numerique_M - U_exact_M, 2) * np.sqrt(h_x)
        liste_erreur_inf.append(erreur_inf_m)
        liste_erreur_2.append(erreur_2_m)
        index += 1
        
    erreur_inf = np.max(liste_erreur_inf)
    erreur_2 = np.max(liste_erreur_2)
    
    
    
    return U_exact_M, U_numerique_M, erreur_2, erreur_inf
```

Dans l'image suivante, on va montrer les erreurs avec les différents valeurs de $\Delta x$ .

```python
##  définir un outil pour donner les images des log d'erreurs avec différents valeurs de h_x
def image_erreur(J_liste, Lambda, type_schema):
    
    X_valeur = 1/J_liste
    Y_valeur_erreur_inf = []
    Y_valeur_erreur_2 = []
    
    for i in np.arange(len(J_liste)):
        h_x, h_t, X, M = espaceDiscretise(J_liste[i], Lambda)
        c_n1, c_0, c_p1 = schema_type(type_schema)
        erreur_2 = schema_erreur(h_x, h_t, X, M, c_n1, c_0, c_p1, J_liste[i])[2]
        erreur_inf = schema_erreur(h_x, h_t, X, M, c_n1, c_0, c_p1, J_liste[i])[3]
        Y_valeur_erreur_2.append(erreur_2)
        Y_valeur_erreur_inf.append(erreur_inf)
        
    log_X_valeur = np.log10(X_valeur)
    log_Y_erreur_inf = np.log10(Y_valeur_erreur_inf)
    log_Y_erreur_2 = np.log10(Y_valeur_erreur_2)
    
    pente_2 = stats.linregress(log_X_valeur, log_Y_erreur_2)[0]
    pente_2 = np.round(pente_2, 2)
    pente_inf = stats.linregress(log_X_valeur, log_Y_erreur_inf)[0]
    pente_inf = np.round(pente_inf, 2)
        
    plt.figure(1)
    fig, ax = plt.subplots(1, 1, figsize=(8,6), dpi = 100)
    ax.set_facecolor('lightyellow')

    
    plt.loglog(X_valeur, Y_valeur_erreur_2, '--', color = 'red',
               label = 'norme $L^2$ pour '+ type_schema +' avec la pente $k_2 =$'+ str(pente_2), linewidth=3)
    plt.loglog(X_valeur, Y_valeur_erreur_inf, color = 'green', 
               label = 'norme $L^{\infty}$ pour '+ type_schema +'avec la pente $k_{inf} =$'+ str(pente_inf), linewidth=3)

    plt.xlabel('différents valeurs de log($\Delta x$)')
    plt.ylabel('log(erreur)')
    plt.legend()
    plt.title( 'Erreur du '+type_schema +' avec différents valeurs de pas $\Delta x$ ')
    plt.show
    
   
```
